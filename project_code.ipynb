__Data Cleaning and EDA__

To prepare the data, we must first consider the various image sizes we have to work with.

```python
import os
from PIL import Image


widths = []
heights = []
# Iterate through each file in the data directory
for subdir, dirs, files in os.walk('./data'):
    for file in files:
        # Open the image file
        with Image.open(os.path.join(subdir, file)) as img:
            width, height = img.size
            widths.append(width)
            heights.append(height)
```

```python
import matplotlib.pyplot as plt

plt.hist(widths,bins=range(0,5000,250))
plt.xlabel('Widths')
plt.ylabel('Frequency')
plt.title('Title')
plt.show()
```

```python
plt.hist(heights,bins=range(0,5000,250))
plt.xlabel('Hieghts')
plt.ylabel('Frequency')
plt.title('Title')
plt.show()
```

We can see there are a wide range of image sizes with most between 0 and 1000 for both width and height.

```python
print('min width:',min(widths),'\nmin height:',min(heights))
print('max width:',max(widths),'\nmax height:',max(heights))
```

For simplicity sake, it might be best to utilize the minimum width and height. This is because the data skews to the right, meaning most of our data lies closest to the minimum as opposed to the maximum. Also, utilizing the minimum values for both width and height ensure that no image will have to be scaled up, and all images will either be already at the 133 x 133 resolution or be rescaled down.

```python
filenames = []
duplicates = []

# Iterate through each file in the data directory
for subdir, dirs, files in os.walk('./data'):
    for file in files:
        # Extract the filename
        filename = os.path.basename(file)

        # append to duplicates if already found
        if filename in filenames:
            duplicates.append(filename)
        else:
            # append to filenames if not
            filenames.append(filename)
```

```python

bars = plt.bar(['Unique Files','Duplicate Files'],[len(filenames), len(duplicates)])
plt.xlabel('Dataset')
plt.ylabel('Number of Duplicates')
plt.title('Title')
plt.legend(bars,['Unique: ' + str(len(filenames)),'Duplicate: ' + str(len(duplicates))])
plt.show()
```

Here we can see that the data contains no duplicate images and we should be able to safely continue with making our models.

## Model building / Model choice

We will begin by building a __Feed Forward Neural Network (FNN)__ model since this can be one of the more simple models to deal with. We can get a baseline for dealing with our train/test sets and iron out any bugs before moving on to more complex learning models.

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# augment data and preprocess (rescale pixel values to be between 0 and 1)
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    './data/train',
    target_size=(133, 133),
    batch_size=24,
    class_mode='binary',
    classes=['dogs','cats']
)

test_generator = test_datagen.flow_from_directory(
    './data/test',
    target_size=(133, 133),
    batch_size=24,
    class_mode='binary',
    classes=['dogs','cats']
)
```

Here we are using the ImageGenerator class from keras to preprocess our images, rescaling the pixel values to be between 0 and 1 to have these values standardized. We use these classes to process our images in batches of 32, rescaling them to 133 x 133 (as determined previously) using the binary class mode since there are only two classes. Additionally, I had to define the classes corresponding to the directories because there were hidden .ipynb files that were being picked up by the .flow_from_directory method, outputting 3 classes, one with 0 images coming from the hidden directory.

Now that we have this processed data, let's recall a batch and show an image using the matplotlib .imshow method.

```python
# Retrieve a batch of images and labels from the train_generator
images, labels = next(train_generator)
# Display the first image from the batch
plt.imshow(images[0])
plt.axis('off')  # Turn off axis
plt.show()

train_generator.reset()
```

We can see that everything is working, so let's proceed to creating our FNN model.

```python
# Define the FNN model
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(133, 133, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

We begin by initializing the sequential model from keras and use the following layers:

1. 2 dimensional convolution layer with 32 3x3 filters using a rectified linear unit activation function. The small kernel size may help with determining edges of shapes that correspond to dogs/cats.

2. 2 x 2 max pooling layer which reduces the spatial dimension of the image

3. Flattening layer that reshapes the output to a vector (this preprocessing is necessary for next layer)

4. Layer that takes a 1 dimensional vector and uses 128 neurons applying the ReLU activation function

5. The output layer with a single neuron using the sigmoid activation function to bring the output to between 0 and 1 for binary classification

We then can compile our model and fit it to our training data:

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Train the model
model.fit(
    train_generator,
    epochs=10,
    validation_data=test_generator
)
```

Here we can see that each epoch fits the model to minimize loss and maximize accuracy on the training data, however the val_loss and val_accuracy are very high/low respectively. We are onl able to get a final accuracy of 0.62 for our final epoch. 

Let's try to tune the learning rate to determine if a different learning rate other than 'Adam' may be more appropriate. Since the default learning rate for the adam optimizer is 0.001, let's try 0.01 and 0.1:

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(0.01), # use learning rate of 0.01
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.fit(
    train_generator,
    epochs=10,
    validation_data=test_generator
)
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.1), # iterate through the three values for lr
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the model
model.fit(
    train_generator,
    epochs=10,
    validation_data=test_generator
)
```

Here we can see that the val_accuracy value, indicating the accuracy of test predictions, decreases as the learning rate is increased, with a test accuracy of only 0.5 with a learning rate of 0.1. This indicates that the model is either randomly guessing or possibly using a single class predictor, which would always be right 50% of the time. This may indicate it is merely memorizing the training data.

Let's try a second model, making a __convolutional neural network (CNN)__ by adding more convolutional layers instead of just one. CNN's should be more suited to image processing than FNN's due to their ability to apply convolutional layers to grid-like data such as image data.

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(133, 133, 3)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(
    train_generator,
    epochs=10,
    validation_data=test_generator
)
```

As you can see, the accuracy is only 61.43% which is close to our initial FNN model, although somewhat less. We can attempt to increase the accuracy further while attempting to tune the hyperparameters by adding a dropout layer with a rate of 0.5 and also adding batch normalization to normalize the activation function.

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(133, 133, 3)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dropout(0.5),  # Adding dropout with a dropout rate of 0.5
    tf.keras.layers.BatchNormalization(),  # Adding batch normalization
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(
    train_generator,
    epochs=10,
    validation_data=test_generator
)
```

We can see this actually lowered the prediction accuracy to 54.19%. This may be due to the dropout layer which is intended to reduce overfitting from the training set due to noise, however this could have also prevented the model from learning from information that proved useful in the initial CNN model.

The issue could also be with the trained images being of low quality, let's use a pretrained model specifically for images and remove the top, fully connected layer. We can use this pre trained model as our first layers in our model and create an additional fully connected layer followed by the sigmoid activation function. I seem to be running out of memory due to this, but let's try reducing the batch size and number of epochs to address this issue.

```python
# augment data and preprocess (rescale pixel values to be between 0 and 1)
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    './data/train',
    target_size=(133, 133),
    batch_size=20,
    class_mode='binary',
    classes=['dogs','cats']
)

test_generator = test_datagen.flow_from_directory(
    './data/test',
    target_size=(133, 133),
    batch_size=20,
    class_mode='binary',
    classes=['dogs','cats']
)
```

```python
from tensorflow.keras.applications import VGG16



pre_trained_model = VGG16(
    weights='imagenet', # pre-trained weights from imagenet
    include_top=False, # do not include fully connected top layers
    input_shape=(133, 133, 3))

# ensure the weights taken from the pre trained model are not updated during training
pre_trained_model.trainable = False

model = tf.keras.Sequential([
    pre_trained_model,
    tf.keras.layers.Flatten(), # necessary for next layer
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(
    train_generator,
    epochs=7,
    validation_data=test_generator
)
```

We can see this pretrained model has an increased performance with the train accuracies lower, but the test accuracies much higher with a final value of 0.7929.
